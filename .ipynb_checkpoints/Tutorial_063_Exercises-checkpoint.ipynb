## Extract the data from time 0 until time 2000
* hint: you can use boolean indexing by selecting the columns from the 'data' matrix that meet: (tx>=0)  & (tx<2000)
* while you're at it, go ahead and grab the timepoints corresponding to that interval as well

```python
ind = (tx>=0)&(tx<2000)
seg_data = data[:,ind]
time = tx[ind]
time.shape
seg_data.shape
```

## Then FFT each trial and take a look at the average power across frequencies
* FFT each trial 
* Then compute the amplitude on each trial
* Then compute the average amplitude across all trials

```python
r,c = seg_data.shape
sp_trial = np.fft.rfft(seg_data, axis =1)
freq_trial = np.fft.rfftfreq(c, 1/sr)
amp = (np.abs(sp_trial)/c)*2
amp_mean = np.mean(amp, axis=0)

plt.plot(freq_trial, amp_mean)
plt.show()
```

## Now design a band pass filter that gets rid of everything outside of the 6-14Hz range (where that big bump is...)
* To design a band pass filter, make a lp filter (with an appropriate fc), then make a hp filter (with an appropriate fc)
* use np.convolve to combine the two filters 
* then call the apply_filt function
* Then you should be left with only those frequencies that pass the lp and the hp filters (i.e. a band pass filter)!

```python
sr = 250
sr1 = 14
sr2 = 6
fc1 = sr1/sr
fc2 = sr2/sr
b = 0.1
window = 'blackman'
type_filt1 = 'lp'
type_filt2 = 'hp'

n1, sinc1 = msf.define_filt(fc1, b, window, type_filt1)
n2, sinc2 = msf.define_filt(fc2, b, window, type_filt2)
bp_filt = np.convolve(sinc1, sinc2)
#bp_filt = sinc1+sinc2
print(bp_filt.shape)

plt.plot(bp_filt)
# plt.plot(n1, sinc2)
# plt.plot(n2, sinc2)
plt.xlabel('Samples', **fig_font)
plt.ylabel('Amplitude', **fig_font)
plt.title('Sinc filter in time domain', **fig_font)
plt.show()

filt_data = np.zeros((r,c))
for i in np.arange(r):
    filt_data[i,:] = msf.apply_filt(input_sig=seg_data[i,:], input_filter=bp_filt)
#print(bp_filt)  

r,c = filt_data.shape
sp_filt = np.fft.rfft(filt_data, axis =1)
freq_filt = np.fft.rfftfreq(c, 1/sr)
amp = (np.abs(sp_filt)/c)*2
amp_mean = np.mean(amp, axis=0)

plt.plot(freq_filt, amp_mean)
plt.show()
```

## You can also make a band-stop filter using a similar logic
* Instead of convolving the lp and the hp filters, you can add them together
* Use this to selectively get rid of the 60Hz line noise in the data by generated a band-stop filter over 50-70Hz

```python
sr1 = 50
sr2 = 70
fc1 = sr1/sr
fc2 = sr2/sr
b1 = 0.1
window = 'blackman'
type_filt1 = 'lp'
type_filt2 = 'hp'

n1, sinc1 = msf.define_filt(fc1, b, window, type_filt1)
n2, sinc2 = msf.define_filt(fc2, b, window, type_filt2)
bs_filt = sinc1+sinc2

plt.plot(n1, sinc2)
plt.plot(n2, sinc2)
plt.xlabel('Samples', **fig_font)
plt.ylabel('Amplitude', **fig_font)
plt.title('Sinc filter in time domain', **fig_font)
plt.show()

filt_data = np.zeros((r,c))
for i in np.arange(r):
    filt_data[i,:] = msf.apply_filt(input_sig=seg_data[i,:], input_filter=bs_filt)
#print(bs_filt)

r,c = filt_data.shape
sp_filt = np.fft.rfft(filt_data, axis =1)
freq_filt = np.fft.rfftfreq(c, 1/sr)
amp = (np.abs(sp_filt)/c)*2
amp_mean = np.mean(amp, axis=0)

plt.plot(freq_filt, amp_mean)
plt.show()

seg_data.shape
```
